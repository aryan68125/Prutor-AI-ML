## How to convert categorical coulmns into numeric columns

### Nominal categorical column

```python
# There is no mathematical order or relationships between values in the categories
```

```python
#For example colors categories
```

```python
df4 = pd.DataFrame([['XL','B',34.5],['L','G',45.12],['S','W',56],['XL','B',26]],columns=['Size','color','price'])
df4
```

```python
#One-hot-encoding:nominal categorical column
df_1 = pd.get_dummies(df4.color)
df_1
```

### Ordinal categorical column

```python
# There is some mathematical order or relationships between categories
```

```python
# For example categories of temperature :->
# Possible categories in temperature high, low, medium
# high > medium > low
```

```python
#Label Encoding:ordinal categorical column
df4['Size'] = df4.Size.map({'XL':3,'L':2,'S':1})
df4
```

```python
df4.drop('color',axis=1,inplace=True)
```

```python
df4
```

```python
## How to concatinate two dataFrames
## Here we are concatenating df_1 and df4 dataFrames into one DataFrame
final_data = pd.concat((df_1,df4),axis=1)
final_data
```

```python
#creating US, India and Others
m = []
for i in df6['native-country']:
    if i == 'United-States':
        m.append("Us")
    elif i == 'India':
        m.append('India')
    elif i == 'England':
        m.append('England')
    else:
        m.append("others")
m
```

```python
df6['native-country'] = m
df6['native-country'].value_counts()
```

```python
d=[]
for i in df6:
    s = df6[i].nunique()
    if s>0:
        d.append(s)
        print('{} has {} categories'.format(i,s))
```
